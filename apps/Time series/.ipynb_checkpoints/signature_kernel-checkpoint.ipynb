# Time series prediction

In this note we consider four different kernels: signature kernel \cite{ASCFLY2021}, signature PDE kernel \cite{ASCFLY2021}, radial basis function kernel, GAK kernel \cite{C2011} and tensornorm kernel(codpy). The tensornorm kernel is based on the tensor technique. We compare above mentioned kernels using mean absolute percentage error (MAPE) (in %) and relative root mean squared error (score) to predict the average price over the next $2$ days given prices over the previous $36$ days. The prices are considered in two different periods: 2017-06-01 - 2018-08-01 and 2020-01-01 - 2021-01-01. 

## Bitcoin price prediction

**Period** 2017-06-01 - 2018-08-01. 

First we reproduce the results in \cite{ASCFLY2021} for the period 2017-06-01 - 2018-08-01. Kwargs describes parameters of the experiment that are passed to our scenario and data generation classes. For the Bitcoin series we do not use signature kernel due to the time it takes to fit its hyperparameters. 


```python
kwargs = {'rescale:xmax': 1000,
'rescale:seed':42,
'sharp_discrepancy:xmax':1000,
'sharp_discrepancy:seed':30,
'sharp_discrepancy:itermax':10,
'discrepancy:xmax':500,
'discrepancy:ymax':500,
'discrepancy:zmax':500,
'discrepancy:nmax':2000,
'validator_compute':['accuracy_score'],
'accuracy_score_function' :  mean_absolute_percentage_error,
'symbols': '',
'initial_date':'2017-06-01',
'finish_date':'2018-08-01',
'h_window':36,
'f_window':2}


scenarios_list = [ (1, -1, -1, -1) ]
scenario_generator_ = SK_scenario_generator()
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), SignKernelPDERegressor(), 
SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "SigKerPDE")
# scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
# SignKernelRegressor(), SK_data_accumulator(), **kwargs)
# scenario_generator_.predictor.plot()
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "rbf")
kwargs['kernel'] = 'gak'
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "gak")
kwargs['codpy_data'] = True
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
codpyexRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "codpy")
```

```python
summary = scenario_generator_.results.dropna(axis=1)
print(summary)
```

**Period 2020-01-01 - 2021-01-01**.

```python
kwargs['codpy_data'] = False
kwargs['initial_date'] = '2020-01-01'
kwargs['finish_date'] = '2021-01-01'
scenario_generator_ = SK_scenario_generator()
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), SignKernelPDERegressor(), 
SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "SigKerPDE")
# scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
# SignKernelRegressor(), SK_data_accumulator(), **kwargs)
# scenario_generator_.predictor.plot()
kwargs['kernel'] = 'rbf'
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "rbf")
kwargs['codpy_data'] = True
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
codpyexRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "codpy")
kwargs['kernel'] = 'gak'
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "gak")
```

```python
summary = scenario_generator_.results.dropna(axis=1)
print(summary)
```

# SP500 price prediction

**Period** 2017-06-01 - 2018-08-01. In this section we use SP500 data downloaded from Yfinance \cite{Yahoo}.

```python
kwargs = {'rescale:xmax': 1000,
'rescale:seed':42,
'sharp_discrepancy:xmax':1000,
'sharp_discrepancy:seed':30,
'sharp_discrepancy:itermax':10,
'discrepancy:xmax':500,
'discrepancy:ymax':500,
'discrepancy:zmax':500,
'discrepancy:nmax':2000,
'validator_compute':['accuracy_score'],
'accuracy_score_function' :  mean_absolute_percentage_error,
'symbols':'^GSPC',
'initial_date':'2017-06-01',
'finish_date':'2018-08-01',
'h_window':36,
'f_window':2}


r = 0.2
scenarios_list = [ (1, -1, -1, -1) ]
scenario_generator_ = SK_scenario_generator()
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), SignKernelPDERegressor(), 
SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "SigKerPDE")
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
SignKernelRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "signature")
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "rbf")
# kwargs['kernel'] = 'gak'
# scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
# RBFtsRegressor(), SK_data_accumulator(), **kwargs)
# scenario_generator_.predictor.plot(name = "gak")
kwargs['codpy_data'] = True
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
codpyexRegressor(), SK_data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "codpy")
```

```python
summary = scenario_generator_.results.dropna(axis=1)
print(summary)
```

**Period 2020-01-01 - 2021-01-01**.

```python
kwargs['initial_date'] = '2020-01-01'
kwargs['finish_date'] = '2021-01-01'
kwargs['codpy_data'] = False
scenario_generator_ = SK_scenario_generator()
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), SignKernelPDERegressor(), 
data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name="SigKerPDE")
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
SignKernelRegressor(), data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name="signature")
kwargs['kernel'] = 'rbf'
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name="rbf")
kwargs['kernel'] = 'gak'
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
RBFtsRegressor(), data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name="gak")
kwargs['codpy_data'] = True
scenario_generator_.run_scenarios(scenarios_list, price_ts_data_generator(), 
codpyexRegressor(), data_accumulator(), **kwargs)
scenario_generator_.predictor.plot(name = "codpy")
```
